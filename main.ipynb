```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

```

```python
folder_path=r'C:\Users\pc\Desktop\projet_de_vente\data-sales-main'
```

```python
#cree un liste pour stocker chaque DataFrame
list_DF=[]
```

```python
# Parcourt chaque fichier dans le dossier
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):  # Vérifie que c'est bien un fichier CSV
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)  # Lis le fichier CSV
        list_DF.append(df)  # Ajoute le DataFrame à la liste
```

```python
# Concatène toutes les DataFrames de la liste en une seule
combined_df = pd.concat(list_DF, ignore_index=True)


# Affiche un aperçu du DataFrame final
print(combined_df.head(5))
```

** En quelle momment doit on faire une compagne publicitaire pour avoir plus de ventes

```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
print(df[['Month','Year','total_sales']].head())
```

```python
df.head(5)
```

```python
df['hour']=df['Order Date'].dt.hour
```

```python
hourly_sales=df.groupby('hour')['total_sales'].sum()
```

```python
plt.figure(figsize=(10, 6))
hourly_sales.plot(kind='bar', color='skyblue')
plt.title('les sales en fonction des heures')
plt.xlabel('heur')
plt.ylabel('profit')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
heurly_commande=df['hour'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
heurly_commande.plot(kind='bar', color='skyblue')
plt.title('les commande en fonction des heures')
plt.xlabel('heur')
plt.ylabel('commande')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
